```python
X=df.iloc[:,:-1]
```

```python
X
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
X.drop(columns=["JoiningYear", "City","Age"], inplace=True)
```

```python
X
```

```python
from sklearn import preprocessing
labelencode= preprocessing.LabelEncoder()
```

```python
X["Education"]=labelencode.fit_transform(X["Education"])
```

```python
X["Gender"]=labelencode.fit_transform(X["Gender"])
X["EverBenched"]=labelencode.fit_transform(X["EverBenched"])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled_xtrain= scaler.fit_transform(X_train)
scaled_xtest= scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("accuracy score of model is {}".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
logmodel= LogisticRegression()
logmodel.fit(scaled_xtrain, y_train)
```

```python
logpredict=logmodel.predict(scaled_xtest)
modelperformance(logpredict)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid={
    "n_neighbors": [3,5,7,9,11],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree","brute"]
}
gridkn=GridSearchCV(KNeighborsClassifier(), param_grid, cv=3)
gridkn.fit(scaled_xtrain, y_train)
```

```python
print("Best Parametres", gridkn.best_params_)
```

```python
knpredict= gridkn.predict(scaled_xtest)
modelperformance(knpredict)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid= {
    "C": [0.1,0.5,1],
    "kernel": ["linear","rbf","poly"]
}
```

```python
gridsvc= GridSearchCV(svm, param_grid, cv=3)
gridsvc.fit(scaled_xtrain,y_train)
```

```python
gridsvc.best_params_
```

```python
predictsvc=gridsvc.predict(scaled_xtest)
modelperformance(predictsvc)
```

```python
from sklearn.tree import DecisionTreeClassifier
param_grid ={
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth": [None, 10, 20,30,4,50],
    "min_samples_split": [2,5,10],
    "min_samples_leaf": [1,2,4]
}
gridsearchtree=GridSearchCV(DecisionTreeClassifier(), param_grid,cv=3)
gridsearchtree.fit(scaled_xtrain,y_train)
```

```python
gridsearchtree.best_params_
```

```python
predicts= gridsearchtree.predict(scaled_xtest)
modelperformance(predicts)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfcmodel= RandomForestClassifier()
n_estimators= [32,64,128, 256]
max_features=[2,3,4]
bootstrap=[True,False]
oob_score=[True,False]
param_grid={ "n_estimators": n_estimators, 
            "max_features": max_features,
            "bootstrap":bootstrap,
            "oob_score": oob_score,
            }
```

```python
gridrfc=GridSearchCV(rfcmodel,param_grid,cv=3)
gridrfc.fit(scaled_xtrain,y_train)
```

```python
gridrfc.best_params_
```

```python
preds=gridrfc.predict(x_test)
modelperformance(preds)
```
